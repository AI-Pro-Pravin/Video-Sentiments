```python
from moviepy.editor import VideoFileClip
import whisper
from transformers import pipeline
```

```python
video_clip = VideoFileClip('sample1.mp4')
audio_clip = video_clip.audio
audio_clip.write_audiofile('sample1.mp3')
```

```python
model = whisper.load_model('base')
```

```python
result = model.transcribe('sample1.mp3',word_timestamps=True)
```

```python
pipe = pipeline(model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
for txt in result['segments']:
  prompt = txt['text']
  timestmp = txt['start']
  resp = pipe(prompt)
  label = resp[0]['label']
  score = resp[0]['score']
  print(f"At {timestmp} '{prompt}' sentiment is {label} with a score of "+"{:.4f}.".format(score))
```
